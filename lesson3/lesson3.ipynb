```python
def fizzbuzz1(n):
    for i in range (1,n+1):
        print(i, end= '')
```

```python
for i in range(0,11):
    fizzbuzz1(i)
    print()
```

# runtime
n = 1, run 1 time

n = 2, run 2 times

...

n = n-1, run n-1 times

O(n) ---> y = x 

```python
def fizzbuzz2(n):
    for i in range (1,n+1):
        for i in range (1, n+1):
            print(i, end= '')
```

```python
for i in range(0,11):
    fizzbuzz2(i)
    print()
```

```python
def fizzbuzz3(n):
    for i in range (1,n+1):
        for i in range (1, n+1):
            for i in range (1, n+1):
                print(i, end= '')
```

```python
for i in range(0,11):
    fizzbuzz3(i)
    print()
```

when n increases, runtime becomes polynomial.
